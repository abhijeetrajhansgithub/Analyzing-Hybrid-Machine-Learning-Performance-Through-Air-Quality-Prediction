### KNN Imputer + Random Forest Model

```python
# prompt: create the prediction code as above but using use  random forest and variable delhi_records_ds_1_copy_rf. train the non NaN target values and then fill the NaN values using the model's prediction of the input features of the espsctive NaN values and name the model model_rf. for any of the features missing in "'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI'", use forward and backward combination of arima and random forest regression and handle "LinAlgError: Schur decomposition solver error." and handle this error as well "ValueError: The feature names should match those that were passed during fit.
# Feature names seen at fit time, yet now missing:
# - AQI"

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Assuming 'delhi_records_ds_1' DataFrame is already loaded and processed.
delhi_records_ds_1_copy_rf = delhi_records_ds_1.copy()

# Define features (X) and target (y)
features = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
target = 'AQI'

# Handle missing values in features using KNNImputer
imputer = KNNImputer(n_neighbors=5)  # You can adjust the number of neighbors
delhi_records_ds_1_copy_rf[features] = imputer.fit_transform(delhi_records_ds_1_copy_rf[features])

# Separate data with and without NaN values in the target
data_with_target = delhi_records_ds_1_copy_rf.dropna(subset=[target])
data_missing_target = delhi_records_ds_1_copy_rf[delhi_records_ds_1_copy_rf[target].isnull()]

# Train Random Forest model on non-NaN target values
X_train = data_with_target[features]
y_train = data_with_target[target]

model_rf = RandomForestRegressor(n_estimators=100, random_state=42) # You can adjust hyperparameters

try:
  model_rf.fit(X_train, y_train)
except ValueError as e:
  if "Feature names seen at fit time, yet now missing" in str(e):
      missing_cols = set(X_train.columns) - set(data_missing_target[features].columns)
      for col in missing_cols:
          data_missing_target[col] = 0  # or handle missing columns more appropriately
      model_rf.fit(X_train, y_train)
  else:
      raise  # Re-raise the exception if it's not the expected ValueError

# Predict AQI for NaN values
X_missing = data_missing_target[features]

try:
    predicted_aqi = model_rf.predict(X_missing)
except ValueError as e:
    if "Feature names seen at fit time, yet now missing" in str(e):
        missing_cols = set(X_train.columns) - set(X_missing.columns)
        for col in missing_cols:
            X_missing[col] = 0
        predicted_aqi = model_rf.predict(X_missing)
    else:
        raise
except np.linalg.LinAlgError as e:
    if "Schur decomposition solver error" in str(e):
      print("Handling LinAlgError: Replacing NaN values with the mean.")
      predicted_aqi = np.nanmean(y_train) # replace with the mean AQI from the training set.
    else:
        raise # Re-raise other LinAlgErrors

# Fill NaN values in the original DataFrame with predictions
data_missing_target[target] = predicted_aqi
delhi_records_ds_1_copy_rf = pd.concat([data_with_target, data_missing_target], ignore_index=True)

# Evaluation metrics
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_train, model_rf.predict(X_train))
r2 = r2_score(y_train, model_rf.predict(X_train))
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2) Score: {r2}")

# Arrange the df according to datetime column
delhi_records_ds_1_copy_rf = delhi_records_ds_1_copy_rf.sort_values(by='Datetime')

delhi_records_ds_1_copy_rf
```

### KNN Imputation + GRU model

```python
# prompt: generate code for knn imputer + gru  in the above format with evaluation metrics. create a copy for the df "delhi_records_ds_1_copy_gru_knn_imp"
# handle all NaN cases in fearures and then train on the non-NaN target AQI then fill the target NaN AQI usinf the model and print the updated df copy

import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense

# Assuming 'delhi_records_ds_1' DataFrame is already loaded.
delhi_records_ds_1_copy_gru_knn_imp = delhi_records_ds_1.copy()

# Define features (X) and target (y)
features = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
target = 'AQI'

# Handle missing values in features using KNNImputer
imputer = KNNImputer(n_neighbors=5)
delhi_records_ds_1_copy_gru_knn_imp[features] = imputer.fit_transform(delhi_records_ds_1_copy_gru_knn_imp[features])

# Normalize features
scaler = MinMaxScaler()
delhi_records_ds_1_copy_gru_knn_imp[features] = scaler.fit_transform(delhi_records_ds_1_copy_gru_knn_imp[features])

# Separate data with and without NaN values in the target
data_with_target = delhi_records_ds_1_copy_gru_knn_imp.dropna(subset=[target])
data_missing_target = delhi_records_ds_1_copy_gru_knn_imp[delhi_records_ds_1_copy_gru_knn_imp[target].isnull()]

# Prepare data for GRU
def create_sequences(data, seq_length):
    xs = []
    ys = []
    for i in range(len(data)-seq_length-1):
        x = data[i:(i+seq_length)]
        y = data[i+seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 10 # Example sequence length
X_train, y_train = create_sequences(data_with_target[features].values, seq_length)
y_train = y_train[:, 0] # Assuming AQI is the first column of the target

# Build GRU model
model = Sequential()
model.add(GRU(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(GRU(units=50))
model.add(Dense(1))  # Output layer for AQI
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)

# Predict AQI for NaN values
X_missing = data_missing_target[features].values
X_missing = X_missing.reshape(X_missing.shape[0], 1, X_missing.shape[1])

predicted_aqi = model.predict(X_missing)
data_missing_target[target] = predicted_aqi

# Concatenate the dataframes
delhi_records_ds_1_copy_gru_knn_imp = pd.concat([data_with_target, data_missing_target], ignore_index=True)

# Evaluate the model
y_pred = model.predict(X_train)
mse = mean_squared_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2) Score: {r2}")

# sort
delhi_records_ds_1_copy_gru_knn_imp = delhi_records_ds_1_copy_gru_knn_imp.sort_values(by='Datetime')

delhi_records_ds_1_copy_gru_knn_imp
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense

# Step 1: Copy the original DataFrame
df_copy = delhi_records_ds_1.copy()

# Step 2: Define features and target
features = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
target = 'AQI'

# Step 3: Impute missing values in features using KNNImputer
imputer = KNNImputer(n_neighbors=5)
df_copy[features] = imputer.fit_transform(df_copy[features])

# Step 4: Normalize features
scaler = MinMaxScaler()
df_copy[features] = scaler.fit_transform(df_copy[features])

# Step 5: Scale target (AQI) values, ensuring non-NaN rows are scaled
target_scaler = MinMaxScaler()
non_nan_target = df_copy[target].dropna().values.reshape(-1, 1)
scaled_target = target_scaler.fit_transform(non_nan_target)

# Replace target with scaled values for non-NaN rows
df_copy['scaled_AQI'] = np.nan
df_copy.loc[df_copy[target].notnull(), 'scaled_AQI'] = scaled_target.flatten()

# Step 6: Split data into training and prediction datasets
train_data = df_copy.dropna(subset=['scaled_AQI'])  # Rows with non-NaN target
predict_data = df_copy[df_copy['scaled_AQI'].isnull()]  # Rows with NaN target

# Step 7: Prepare sequences for GRU model training
def create_sequences(data, target, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length]
        y = target[i + seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 10
X_data = train_data[features].values
y_data = train_data['scaled_AQI'].values

if len(X_data) > seq_length:
    X_train, y_train = create_sequences(X_data, y_data, seq_length)
else:
    raise ValueError("Insufficient data for creating sequences. Increase your dataset size.")

# Step 8: Train the GRU model
model = Sequential([
    GRU(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    GRU(units=50),
    Dense(1)  # Output layer for AQI
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)

# Step 9: Replace NaN values in the target column in-place
for index, row in df_copy.iterrows():
    if pd.isna(row[target]):  # Check if target is NaN
        # Prepare input features
        input_features = row[features].values.astype(float).reshape(1, 1, -1)  # Shape: (1, 1, num_features)

        # Predict scaled value and inverse transform to original scale
        predicted_scaled = model.predict(input_features, verbose=0)
        predicted_original = target_scaler.inverse_transform(predicted_scaled)

        # Replace NaN with predicted value
        df_copy.at[index, target] = predicted_original.flatten()[0]

# Step 10: Restore original feature and target values
df_copy[features] = scaler.inverse_transform(df_copy[features])
df_copy.drop(columns=['scaled_AQI'], inplace=True)

# Step 11: Validate the target column
if df_copy[target].isnull().any():
    print("Error: Some NaN values remain in the target column.")
else:
    print("All NaN values in the target column have been successfully replaced.")

# Step 12: Sort by Datetime
df_copy = df_copy.sort_values(by='Datetime')

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Step 13: Evaluation Metrics

# Predict on training data to evaluate the model
y_pred = model.predict(X_train)

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y_train, y_pred))
print(f"RMSE: {rmse}")

# Calculate MAE (Mean Absolute Error)
mae = mean_absolute_error(y_train, y_pred)
print(f"MAE: {mae}")

# Calculate MSE (Mean Square Error)
mse = mean_squared_error(y_train, y_pred)
print(f"MSE: {mse}")

# Calculate R² (Coefficient of Determination)
r2 = r2_score(y_train, y_pred)
print(f"R²: {r2}")

# Final DataFrame
print(df_copy)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense

# Step 1: Copy the original DataFrame
df_copy = delhi_records_ds_1.copy()

# Step 2: Define features and target
features = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
target = 'AQI'

# Step 3: Impute missing values in features using spline interpolation
df_copy[features] = df_copy[features].interpolate(method='spline', order=3, limit_direction='both')

# Step 4: Normalize features
scaler = MinMaxScaler()
df_copy[features] = scaler.fit_transform(df_copy[features])

# Step 5: Scale target (AQI) values, ensuring non-NaN rows are scaled
target_scaler = MinMaxScaler()
non_nan_target = df_copy[target].dropna().values.reshape(-1, 1)
scaled_target = target_scaler.fit_transform(non_nan_target)

# Replace target with scaled values for non-NaN rows
df_copy['scaled_AQI'] = np.nan
df_copy.loc[df_copy[target].notnull(), 'scaled_AQI'] = scaled_target.flatten()

# Step 6: Split data into training and prediction datasets
train_data = df_copy.dropna(subset=['scaled_AQI'])  # Rows with non-NaN target
predict_data = df_copy[df_copy['scaled_AQI'].isnull()]  # Rows with NaN target

# Step 7: Prepare sequences for LSTM model training
def create_sequences(data, target, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length]
        y = target[i + seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 10
X_data = train_data[features].values
y_data = train_data['scaled_AQI'].values

if len(X_data) > seq_length:
    X_train, y_train = create_sequences(X_data, y_data, seq_length)
else:
    raise ValueError("Insufficient data for creating sequences. Increase your dataset size.")

# Step 8: Train the 4-layered Bidirectional LSTM model
model = Sequential([
    Bidirectional(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]))),
    Bidirectional(LSTM(units=50, return_sequences=True)),
    Bidirectional(LSTM(units=50, return_sequences=True)),
    Bidirectional(LSTM(units=50)),
    Dense(1)  # Output layer for AQI
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)

# Step 9: Replace NaN values in the target column in-place
for index, row in df_copy.iterrows():
    if pd.isna(row[target]):  # Check if target is NaN
        # Prepare input features
        input_features = row[features].values.astype(float).reshape(1, 1, -1)  # Shape: (1, 1, num_features)

        # Predict scaled value and inverse transform to original scale
        predicted_scaled = model.predict(input_features, verbose=0)
        predicted_original = target_scaler.inverse_transform(predicted_scaled)

        # Replace NaN with predicted value
        df_copy.at[index, target] = predicted_original.flatten()[0]

# Step 10: Restore original feature and target values
df_copy[features] = scaler.inverse_transform(df_copy[features])
df_copy.drop(columns=['scaled_AQI'], inplace=True)

# Step 11: Validate the target column
if df_copy[target].isnull().any():
    print("Error: Some NaN values remain in the target column.")
else:
    print("All NaN values in the target column have been successfully replaced.")

# Step 12: Sort by Datetime
df_copy = df_copy.sort_values(by='Datetime')

# Step 13: Evaluation Metrics

# Predict on training data to evaluate the model
y_pred = model.predict(X_train)

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y_train, y_pred))
print(f"RMSE: {rmse}")

# Calculate MAE (Mean Absolute Error)
mae = mean_absolute_error(y_train, y_pred)
print(f"MAE: {mae}")

# Calculate MSE (Mean Square Error)
mse = mean_squared_error(y_train, y_pred)
print(f"MSE: {mse}")

# Calculate R² (Coefficient of Determination)
r2 = r2_score(y_train, y_pred)
print(f"R²: {r2}")

# Final DataFrame
print(df_copy)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Step 1: Copy the original DataFrame
df_copy = delhi_records_ds_1.copy()

# Step 2: Define features and target
features = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
target = 'AQI'

# Step 3: Impute missing values in features using EMA
for feature in features:
    df_copy[feature] = df_copy[feature].ewm(span=5, adjust=False).mean()  # Exponential Moving Average (span=5)

# Step 4: Normalize features
scaler = MinMaxScaler()
df_copy[features] = scaler.fit_transform(df_copy[features])

# Step 5: Scale target (AQI) values, ensuring non-NaN rows are scaled
target_scaler = MinMaxScaler()
non_nan_target = df_copy[target].dropna().values.reshape(-1, 1)
scaled_target = target_scaler.fit_transform(non_nan_target)

# Replace target with scaled values for non-NaN rows
df_copy['scaled_AQI'] = np.nan
df_copy.loc[df_copy[target].notnull(), 'scaled_AQI'] = scaled_target.flatten()

# Step 6: Split data into training and prediction datasets
train_data = df_copy.dropna(subset=['scaled_AQI'])  # Rows with non-NaN target
predict_data = df_copy[df_copy['scaled_AQI'].isnull()]  # Rows with NaN target

# Step 7: Prepare sequences for LSTM model training
def create_sequences(data, target, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i + seq_length]
        y = target[i + seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 10
X_data = train_data[features].values
y_data = train_data['scaled_AQI'].values

if len(X_data) > seq_length:
    X_train, y_train = create_sequences(X_data, y_data, seq_length)
else:
    raise ValueError("Insufficient data for creating sequences. Increase your dataset size.")

# Step 8: Train the 4-layered Bidirectional LSTM model
model = Sequential([
    Bidirectional(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]))),
    Bidirectional(LSTM(units=50, return_sequences=True)),
    Bidirectional(LSTM(units=50, return_sequences=True)),
    Bidirectional(LSTM(units=50)),
    Dense(1)  # Output layer for AQI
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)

# Step 9: Replace NaN values in the target column in-place
for index, row in df_copy.iterrows():
    if pd.isna(row[target]):  # Check if target is NaN
        # Prepare input features
        input_features = row[features].values.astype(float).reshape(1, 1, -1)  # Shape: (1, 1, num_features)

        # Predict scaled value and inverse transform to original scale
        predicted_scaled = model.predict(input_features, verbose=0)
        predicted_original = target_scaler.inverse_transform(predicted_scaled)

        # Replace NaN with predicted value
        df_copy.at[index, target] = predicted_original.flatten()[0]

# Step 10: Restore original feature and target values
df_copy[features] = scaler.inverse_transform(df_copy[features])
df_copy.drop(columns=['scaled_AQI'], inplace=True)

# Step 11: Validate the target column
if df_copy[target].isnull().any():
    print("Error: Some NaN values remain in the target column.")
else:
    print("All NaN values in the target column have been successfully replaced.")

# Step 12: Sort by Datetime
df_copy = df_copy.sort_values(by='Datetime')

# Step 13: Evaluation Metrics

# Predict on training data to evaluate the model
y_pred = model.predict(X_train)

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y_train, y_pred))
print(f"RMSE: {rmse}")

# Calculate MAE (Mean Absolute Error)
mae = mean_absolute_error(y_train, y_pred)
print(f"MAE: {mae}")

# Calculate MSE (Mean Square Error)
mse = mean_squared_error(y_train, y_pred)
print(f"MSE: {mse}")

# Calculate R² (Coefficient of Determination)
r2 = r2_score(y_train, y_pred)
print(f"R²: {r2}")

# Final DataFrame
print(df_copy)
```

```python
# prompt: conduct Outlier detection and treatment

# Outlier detection using IQR method
def treat_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower_bound, upper_bound)
    return df

# Treat outliers for each pollutant column
for pollutant in pollutants:
    delhi_records_ds_1 = treat_outliers_iqr(delhi_records_ds_1, pollutant)

# You can explore other outlier detection methods as well, such as the z-score method.

# Example of Z-score method (optional)
# from scipy import stats
# for pollutant in pollutants:
#     z = np.abs(stats.zscore(delhi_records_ds_1[pollutant]))
#     delhi_records_ds_1 = delhi_records_ds_1[(z < 3)] # Removing outliers beyond 3 standard deviations

# After outlier treatment, re-plot distributions to see the effect:
plt.figure(figsize=(15, 10))
for i, pollutant in enumerate(pollutants):
    plt.subplot(4, 3, i+1)
    sns.histplot(delhi_records_ds_1[pollutant], kde=True)
    plt.title(f'Distribution of {pollutant} (After Outlier Treatment)')
plt.tight_layout()
plt.show()
```
